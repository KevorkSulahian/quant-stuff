## **Skewness** 
Skewness is a statistical metric which attempts to measure the asymmetry of a particular distribution under consideraton (Turney, 2022). If the right side of a given distribution is not a mirror image of the left side then the distribution is asymmetrical. When describing the skewness of a distribution researchers usually consider three (3) different scenarios. These scenarios are negative (left), positive (right) as well as a zero skewed distribution. 
### Negative (left) Skewness.
When a distribution is negatively skewed or skewed to the left implies that the distribution is longer on the left side of its peak. In a negatively skewed distribution it can be observed that the smaller values in the left tail are much longer as compared to the larger values in the right. This is depicted in fig 1 in page (3) indicating all the three types of distribution. It should also be observed that when a distribution is negatively skewed, the median is greater than the mean and the median is usually less than the mode. 


```python
#@title Left Skew data
left_skewed_data = np.random.binomial(10, 0.2, 1000)

print("Left-Skewed Dataset")
print("Mean:", np.mean(left_skewed_data))
print("Median:", np.median(left_skewed_data))

plt.hist(left_skewed_data, bins=20, color='blue')
plt.xlabel('Data Values')
plt.ylabel('Frequency')
plt.title('Left-Skewed Dataset')
plt.show()
```

### Positive (Right) Skewed Distribution.   
On the other hand, a distribution can be positively skewed. When a distribution is positively skewed, it implies that the particular distribution is much longer on its right side of the peak as compared to its left side. In a positively skewed distribution, the mode is usually less than the median and mean greater than the median. The reason why the mean of a positively skewed distribution is usually greater than the median is due to extreme values within the tail which affect the mean more than the median. 



```python
#@title Right Skew data
right_skewed_data = np.random.binomial(10, 0.8, 1000)

print("\nRight-Skewed Dataset")
print("Mean:", np.mean(right_skewed_data))
print("Median:", np.median(right_skewed_data))

plt.hist(right_skewed_data, bins=20, color='red')
plt.xlabel('Data Values')
plt.ylabel('Frequency')
plt.title('Right-Skewed Dataset')
plt.show()
```

## Zero Skewness
When a distribution has zero skew, it implies that the distribution under examination is symmetrical. This means that the two sides (Right and left) are considered to be mirror images. Statistically, normally distributed data has zero skew and in addition to that uniform distribution as well as bimodal also known as two-peak distribution they also have zero skew.

```python
#@title 0 Skew data
right_skewed_data = np.random.binomial(10, 0.5, 1000)

print("\n Non-Skewed Dataset")
print("Mean:", np.mean(right_skewed_data))
print("Median:", np.median(right_skewed_data))

plt.hist(right_skewed_data, bins=20, color='green')
plt.xlabel('Data Values')
plt.ylabel('Frequency')
plt.title('Right-Skewed Dataset')
plt.show()
```

### **Impact of skewed data on linear regression** 
The presence of skewness in any distribution is undesirable as it results in the problem of outliers (Sharma, 2022). It has been observed that the tail region of a skewed distribution usually acts as a source of outliers for a statistical model. Usually the Ordinary Least Square method performs badly in the presence of outliers (Gujarati, 2004). 


### **How do we handle skewed distribution?** 
If researchers detect the presence of skewness in a distribution, then there is a need for the data to be transformed to a  Gaussian or a normally distributed variable. A skewed distribution can be transformed to a Gaussian distribution through log transformation. It involves the use of a logarithmic function on skewed data. Each value is replaced with its logarithm. A log transformation can help to fit a very skewed distribution into a Gaussian one.


### **How to detect Skewness in a distribution.**
It can be observed that there are several formulas that are employed to measure skewness (Sharma, 2022). However, the most widely employed measure of skewness is Pearson's median skewness. It is credited for taking advantage of the discrepancies between the mean and median in a particular skewed distribution. It is calculated as follows: 


<h3 align="center">Pearson's median Skewness = 3 * (Mean -Median)/ Standard Deviation.</h3> 


Pearson's Median Skewness informs us how many standard deviations do separate the mean and median. Histograms can also be used to detect the presence of skewness in a distribution. The distribution which is used to come up with a histogram can reveal the connection between the mean, median and mode of the distribution. The position of these measures of central tendency shapes the histogram and hence reveal if the distribution is negatively, positively and normally skewed.  








## **Non-stationarity**
Researchers usually make use of time series data from the field of finance and economics which include asset prices, Gross domestic product, exchange rates as well as other macro-economic indicators with the main objective of forecasting economic movements and also analyzing markets (Lordanova., et al, 2022). It has also been observed that financial and economic time series data usually exhibits the problem of non-stationarity (Muma & Maphosa, 2017).   
A series is considered to be non- stationary as long as its mean does not persistently goes back to its mean and in addition does have the propensity to have a drift, then that particular series is considered to be non-stationary ( Paramaiah & Akway, 2008). However, if the mean as well as the variance of a time series are consistently constant overtime, and that the covariance’s value across the two periods relies on the gap across the periods and not on the authentic time with which the covariance is taken into account, this therefore implies that the times series is stationary. 


```python
#@title non-stationary vs stationary data


stationary_data = np.random.normal(loc=0, scale=1, size=1000)
nonstationary_data = np.random.normal(loc=0, scale=np.linspace(1, 2, 1000), size=1000)


fig, axs = plt.subplots(1, 2, figsize=(10, 5))

axs[0].plot(stationary_data, color='blue')
axs[0].set_xlabel('Time')
axs[0].set_ylabel('Data Values')
axs[0].set_title('Stationary Dataset')

axs[1].plot(nonstationary_data, color='red')
axs[1].set_xlabel('Time')
axs[1].set_ylabel('Data Values')
axs[1].set_title('Non-Stationary Dataset')

plt.show()
```

### **Effect of the presence of non-stationary series**

Townsend, (2001) observed that it is critical for researchers to determine the order of integration prior to the application of econometric methods. He further suggested that if variables are of the wrong order of integration then the outcome of such an analysis is most likely to result into spurious regression results. 
Time series studies have indicated that the order of integration is a critical component of the data characteristics hence there is need for researchers to determine the correct order of integration prior to any econometric applications. As a result of a wrong order ofintegration, 
the results are a spurious regression which may also be refered to as wrong test stattistics. Spurious regression results makes the whole analysis useless (Gujarati, 2004). In addition, if the wrong order of integration is applied then co-integration tests would also be unreliable (Muma & Maposa, 2017). 

How Do we handle non-stationarity? 

There are several ways in which non-stationary time series can be convertered into a stationary series. One of the most common ways include differencing. When a time series is differenced, it is refered to as being integrated and would be denoted by I(d). In this case, d represents the order of integration. It is also important to note that the order of integration represents the unit roots which are in a particular series. some other researchers noted that it may also denote the frequency of differencing operations which are needed to make a time series stationary.   

When a time series is not stationary, it can be converted into a stationary series by differencing among other ways. A different stationary series is said to be integrated and is denoted as I(d) where d is the order of integration. The order of integration is the number of unit roots contained in the series, or the number of differencing operations it takes to make the series stationary.


**How to detect the presence of unit root in a time series?**

When a time series shows a trend or seasonal effects, then the researcher need to underastand that the series is non-stationary. Another formal way of determining if a series is non-stationaary is through the use of a summary statistics. By dividing the data set into portions, then compare the mean and variance of each group. If the differences of these measures are statistically significant then the series is non-stationary.


## **Sensitivity to outliers**

Now according to (Grubbs 1969) An outlier is an observation “that appears to deviate markedly from other members of the sample in which it occurs”. Now when it comes to Machine Learning, It is simply defined as a point in the dataset that is far away from the other points, most of the time it is considered as an error or an improbable point.



```python
#@title Outlier data

x = np.random.normal(50, 10, 50)
y = np.random.normal(160, 20, 50)

x[3] = 10
y[10] = 220
x[25] = 90
y[35] = 100

plt.scatter(x, y)
outliers = [3, 10, 25, 35]
plt.scatter(x[outliers], y[outliers], facecolor='none', edgecolor='r', s=100)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Example Scatter Plot with Outliers')
plt.show()
```

As seen in the image above, there are outliers which are pointed out in the graph.
Now that we have our outliers, we will return to how to identify them in a bit. We also have to understand the source of these outliers, for example the one on the far left and the one where the age is 50 are probably mistakes, or wrong data. While the other two are likely extreme values.
But when it comes to detecting outliers, there are several ways to do so including but not limited to, using IQR, Z-scores, Winsorization, T-stat, cook’s distance,  and others.

### IQR
With IQR, first we will need to calculate 


<h3 align="center">Q1 and Q3 to get IQR = Q3 - Q1</h3> 

And then data points that fall below

<h3 align="center">Q1 - 1.5 * IQR</h3> 

or points that are above

<h3 align="center">Q3 + 1.5 * IQR</h3> 

  would be considered as outliers.


```python
#@title Outlier vs without outlier data
def remove_outlier_IQR(df):
    Q1=df.quantile(0.25)
    Q3=df.quantile(0.75)
    IQR=Q3-Q1
    df_final=df[~((df<(Q1-1.5*IQR)) | (df>(Q3+1.5*IQR)))]
    return df_final

x_removed = remove_outlier_IQR(pd.DataFrame(x))[0]

for i in np.where(np.isnan(x_removed))[0]:
  x2 = np.delete(x,i)
  y2 = np.delete(y,i)

y_removed = remove_outlier_IQR(pd.DataFrame(y))[0]

for j in np.where(np.isnan(x_removed))[0]:
  x2 = np.delete(x2,j)
  y2 = np.delete(y2,j)

fig, axs = plt.subplots(1, 2, figsize=(10, 5))

axs[0].scatter(x, y, color='blue')
axs[0].set_xlabel('X')
axs[0].set_ylabel('Y')
axs[0].set_title('Original Data')

axs[1].scatter(x2, y2, color='red')
axs[1].set_xlabel('X')
axs[1].set_ylabel('Y')
axs[1].set_title('Filtered Data')

# Show the plot
plt.show()
```

# **Kurtosis / Heteroscedasticity**
# Definition :
Kurtosis is a measure of the degree to which the tails of a distribution differs from a normal distribution.

Kurtosis = (1/n) * SUM[(xi - mean)^4] / (std dev)^4

Heteroscedasticity is the unstable friction of residuals across the range of a predictor variable.

Var(εi) = σi^2

# Description :
In finance, two important generalities are kurtosis and heteroscedasticity. Kurtosis is used to measure the degree of threat or volatility in a fiscal asset's returns by assaying the shape of its distribution. A high kurtosis value indicates that the asset's returns are more unpredictable and may have further extreme issues. Heteroscedasticity, on the other hand, occurs when the variability of a fiscal asset's returns changes over time, rather than remaining constant. This can have significant counter accusations for fiscal modeling and threat operation as it can lead to inaccurate prognostications and mispricings in fiscal requests. Understanding these generalities is pivotal for anyone working in finance to make informed investment opinions, develop effective threat operation strategies, and produce accurate fiscal models.



# Demonstration :
In the field of finance and investing, high kurtosis in returns implies that there have been multitudinous significant price oscillations in history, swinging from the average returns for the investment. An investment with high kurtosis is likely to witness extreme price oscillations, and therefore requires close attention from our trading office. As we touched upon the conception of heteroscedasticity in the former challenge, we will now concentrate on the implicit threat of kurtosis

```python
np.random.seed(0)
mu, sigma = 0, 1
no_of_obs = 252
x = np.random.normal(mu, sigma, no_of_obs)
sigma = np.where(x < 0, 2, 1) * sigma
x_h = np.random.normal(mu, sigma, no_of_obs)
print('Random Sample data')
print(x_h[0:10])
```

# Diagram : 

```python
#@title Heteroscedastic Time Series Data Plot

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(x, label='Original Data')
ax.plot(x_h, label='Heteroscedastic Data')
ax.set(title='Heteroscedastic Time Series Data',
       xlabel='Time',
       ylabel='Value')
ax.legend(loc='upper left')
plt.show()
```

# Diagnosis :
We can diagnose the presence of kurtosis/ heteroscedasticity in a time series data by conniving the data and observing if there are any extreme values or if there's unstable variability in the data over time. Also, statistical tests like the Jarque- Bera test can be used to describe the presence of kurtosis in the data.

# Damage :
Kurtosis/ heteroscedasticity can beget significant damage to the delicacy of the volatility model as it assumes that the distribution of returns is normal and has constant friction overtime.However, the volatility model can be inaccurate, leading to poor pricing and hedging opinions, If the data violates these hypotheticals.

# Directions :
To deal with the issue of kurtosis/heteroscedasticity in the data, we have the option to use models such as GARCH (Generalized Autoregressive Conditional Heteroskedasticity) or EGARCH (Exponential GARCH), which consider the non-uniform variability in the data. These models allow for the variability in the data to change over time and take into account the existence of outliers in the data. Alternatively, we can use robust statistical techniques like M-estimation, which can reduce the effect of outliers in the data.
